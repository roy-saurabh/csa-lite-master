# 01 Dataset Overview

This notebook provides an overview of the CSA-lite case dataset.

**Disclaimer:** CSA-lite does not produce legal classifications, compliance
determinations, conformity assessments, or validated harm predictions.
The Context Severity Index is a structured-coding output, not a legal risk class.

```python
from pathlib import Path
import pandas as pd

from csalite.io import read_cases_csv
from csalite.validation import validate_dataframe
from csalite.scoring import score_dataframe

DATA_DIR = Path('../data/processed')
cases_path = DATA_DIR / 'csa_lite_cases.csv'
```

```python
# Load and validate cases
df = read_cases_csv(cases_path)
print(f'Loaded {len(df)} cases')
result = validate_dataframe(df)
print(result.summary())
```

```python
# Score the dataset
scored = score_dataframe(df)
scored.head()
```

```python
# Cases by Annex III area
scored['annex_iii_area'].value_counts().sort_index()
```

```python
# Severity band distribution
scored['context_severity_band'].value_counts()
```

```python
# Evidence completeness summary
scored['evidence_completeness_index'].describe()
```
