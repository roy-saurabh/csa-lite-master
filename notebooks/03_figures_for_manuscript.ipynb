# 03 Figures for Manuscript

This notebook generates all 8 manuscript figures using the CSA-lite plots module.

**Disclaimer:** CSA-lite does not produce legal classifications.
The Context Severity Index is a structured-coding output, not a legal risk class.

```python
from pathlib import Path
import matplotlib.pyplot as plt

from csalite.io import read_cases_csv
from csalite.scoring import score_dataframe
from csalite.plots import generate_all_figures

DATA_DIR = Path('../data/processed')
FIGURES_DIR = Path('../outputs/figures')
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

df = read_cases_csv(DATA_DIR / 'csa_lite_cases.csv')
scored = score_dataframe(df)
print(f'{len(scored)} cases loaded and scored')
```

```python
# Generate all figures (saved to outputs/figures/)
figs = generate_all_figures(scored, FIGURES_DIR)
print(f'Generated {len(figs)} figures')
```

```python
# Display Fig 1 (pipeline architecture) inline
from csalite.plots import fig_1_pipeline_architecture
fig_1_pipeline_architecture(FIGURES_DIR)
plt.show()
```

```python
# Display Fig 3 (CSI boxplot by Annex area) inline
from csalite.plots import fig_3_context_severity_by_annex_area
fig_3_context_severity_by_annex_area(scored, FIGURES_DIR)
plt.show()
```
