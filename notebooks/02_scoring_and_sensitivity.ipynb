# 02 Scoring and Sensitivity Analysis

This notebook demonstrates the CSA-lite scoring methodology and
sensitivity analysis under three null-imputation assumptions.

**Disclaimer:** CSA-lite does not produce legal classifications.
The Context Severity Index is a structured-coding output, not a legal risk class.

```python
from pathlib import Path
import pandas as pd

from csalite.io import read_cases_csv
from csalite.scoring import score_dataframe
from csalite.sensitivity import compute_sensitivity_dataframe, compute_band_change_summary

DATA_DIR = Path('../data/processed')
df = read_cases_csv(DATA_DIR / 'csa_lite_cases.csv')
scored = score_dataframe(df)
```

```python
# Sensitivity analysis
sens = compute_sensitivity_dataframe(scored)
sens[['case_id', 'case_name',
      'unknown_as_zero_band', 'unknown_as_neutral_band', 'unknown_as_conservative_band',
      'sensitivity_band_change', 'missing_dimensions_count']].head(20)
```

```python
# Band change summary by area
compute_band_change_summary(sens)
```
